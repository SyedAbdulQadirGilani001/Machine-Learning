```python
from itertools import combinations # itertools for combinations

def main(): # main function
  word = 'bath' # word for combinations
  for string in combinations(word, 3): # combinations of word
    print(''.join(string)) # print combinations

if __name__ == '__main__': # if main
  main() # run main
```
