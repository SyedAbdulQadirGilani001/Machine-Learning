```python
import csv
from tiktokapi import TikTokAPI

def scrape_influencers_csv(min_followers, max_followers, country_code=None):
    api = TikTokAPI()
    influencers = []

    while len(influencers) < 3000:
        results = api.search_for_you_page(
            count=100,
            country_code=country_code
        )
        
        for tiktok in results.get("items", []):
            user = tiktok.get("author", {})
            follower_count = user.get("followers_count", 0)
            
            if min_followers <= follower_count <= max_followers:
                username = user.get("unique_id", "")
                name = user.get("nickname", "")
                email = ""  # You'll need to find another way to obtain the email
                profile_url = f"https://www.tiktok.com/@{username}"
                influencers.append({
                    "Name": name,
                    "Email": email,
                    "Number of followers": follower_count,
                    "TikTok username": username,
                    "TikTok profile URL": profile_url
                })

    return influencers

# Scrape influencers with 5k-10k followers interested in lyrics or music
lyrics_music_influencers = scrape_influencers_csv(5000, 10000)

# Save the influencers data to a CSV file
with open("lyrics_music_influencers.csv", "w", newline="") as file:
    fieldnames = lyrics_music_influencers[0].keys()
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(lyrics_music_influencers)
```

```python
# 2
import numpy as np # for linear algebra

d = 2 # dimension of the array

array = np.arange(1, 16) # create an array of integers from 1 to 15
array = array.reshape(3, 5) # reshape the array to have 3 rows and 5 columns

print(array[d - 1, :],'\n') # Select row d - 1

print(array[:, d],'\n') # Select column d

print(array[[0, 1], :],'\n') # Select rows 0 and 1

print(array[:, 2:],'\n') # Select columns 2-d

print(array[1, d],'\n') # Select the element that is in row 1 and column d

print(array[1:3, [0, 2, d]],'\n') # Select rows 1 and 2 and columns 0, 2, and d
```
